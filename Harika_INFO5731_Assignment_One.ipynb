# Question 1

(20 points). Write a Python program to generate 12 random numbers between 1 and 100. Sort the list of numbers in ascending order and print it out. Sort the list of numbers in descending order and print it out. Calculate the average for the group and print it out.

```
# Your code here
import random

#Using function to generate desired no. of random numbers
def random_no_generator(required_no):
  return random.sample(range(1,100),required_no)

#Using function to sort the random numbers list in ascending order
def sort_ascend(random_list):
  random_list.sort()
  return random_list

#Using function to sort the random numbers list in descending order
def sort_descend(random_list):
  return sorted(random_list,reverse=True)

#Using function to calculate average of the random numbers
def average(number_list):
  sum = 0
  for each in number_list:
    sum +=each
  return sum/len(number_list)

#Initialising 12 random numbers using above function
rand_no_list = random_no_generator(12)

#Printing 12 random numbers using above function
print("Question no.: 1.(a)")
print("12 Random numbers from 1 to 100: ", rand_no_list)

#Sorting and printing ascending ordered results using above function
print("Question no.: 1.(b)")
print("Ascending order: ", sort_ascend(rand_no_list))

#Sorting and printing descending ordered results using above function
print("Question no.: 1.(c)")
print("Descending order: ", sort_descend(rand_no_list))

#Calculating and printing results using above function
print("Question no.: 1.(d)")
print("Average of these random numbers: ", average(rand_no_list))
```

# Question 2
(25 points). Write a program that will do the following string formatting tasks:

(1) Ask the user to enter two sentences.

(2) Make each sentence into a list. Each element in the list should correspond to a word in the sentence. For example, if the sentence is "Alas! Am I alive? This is my first python program", the program should print out ['Alas, '!', 'Am', 'I', 'alive, '?', 'This', 'is', 'my', 'first', 'python', 'program'].

(3) Compare the two sentences. Print out a list of words that occur in both sentences.

(4) Remove any punctuation from the sentence and print it back out. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”)

(5) Count the number of vowels in the sentences. Print out each vowel and the number of times it appears in the sentences, such as: a:2, e:1, i:0, etc

```
# Your code here

import string, nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

#Using function to split two sentences into list of words
def tokenizedSentence(sentence_1,sentence_2):
  return (word_tokenize(sentence_1),word_tokenize(sentence_2))

#Using function to find common words between two sentences
def common_words(list_1,list_2):
  common_words_list = []
  for each in list_1:
    if each in list_2 and each not in common_words_list:
      common_words_list.append(each)
  return common_words_list if common_words_list else "There are no common words"

#Using function to remove punctuation from gievn sentence
def removingPunctuation(samp_sentence):
  table = str.maketrans(dict.fromkeys(string.punctuation))
  return samp_sentence.translate(table)

#Using function to count vowel occurrences in a sentence
def vowel_occurrence(samp_sentence):
  vowels = {each:samp_sentence.lower().count(each) for each in 'aeiou'}
  return vowels

#Using function to take input from user
def user_input():
  samp_sent_1 = input("Please enter any sentence: ")
  samp_sent_2 = input("Please enter another sentence: ")
  return (samp_sent_1,samp_sent_2)

#Calling above function to take user input and printing results
sentences = user_input()
print("\nQuestion no.: 2.(a)")
print("Your two input sentences are : \n", sentences[0],sentences[1])

#Calling above function to split sentences and printing results 
result_tuple = tokenizedSentence(sentences[0],sentences[1])
print("\nQuestion no.: 2.(b)")
print("Tokenized words from sentence 1: ", result_tuple[0])
print("Tokenized words from sentence 2: ", result_tuple[1])

#Calling above function to find and print common words in two sentences 
print("\nQuestion no.: 2.(c)")
print("Common words from above two sentences are : ",common_words(result_tuple[0],result_tuple[1]))

#Calling above function to remove punctuation in a sentence and printing results 
print("\nQuestion no.: 2.(d)")
print("Removing punctuations from sentence 1: \n ",removingPunctuation(sentences[0]))
print("Removing punctuations from sentence 2: \n ",removingPunctuation(sentences[1]))

#Calling above function to count vowel occurrences in a sentence and printing results 
print("\nQuestion no.: 2.(e)")
print("Vowel occurrences in sentence 1 are : ",vowel_occurrence(sentences[0]))
print("Vowel occurrences in sentence 2 are : ",vowel_occurrence(sentences[1]))
```

(15 points). The formula that describes the volume V of a sphere with radius r is the following:

$ V=\frac{4}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of 1–10. Output the result in the following format:

	r	V
	1	…
	2	…
	3	…


```
# Your code here

#Using function to calculate Volume of Sphere for range of r values and printing in given format
def volumeOfSphere():
  print("\nQuestion no.: 3")
  print('\nr\tV')
  for radius in range(1,11):
    volume = (4/3)*(3.14)*radius*radius*radius
    print('\n'+str(radius)+'\t'+str(volume))
  return

#Calling above function to calculate and print results in given format 
volumeOfSphere()

```

```
# Your code here
import csv,os,math
from google.colab import drive,files

'''Using function to open csv file and calculate consecutive year Occupied homes differences 
and printing results in given format'''
def diff_calc():

  #Storing and opening csv file from gdrive 
  drive.mount('/content/gdrive')
  file_list = os.listdir('/content/gdrive/My Drive/Colab Notebooks')
  os.chdir('/content/gdrive/My Drive/Colab Notebooks')

  #Reading csv file contents
  if 'Assignment1_denton_housing.csv' in file_list:
    with open('Assignment1_denton_housing.csv') as csv_file:
      csv_reader = csv.reader(csv_file, delimiter=',')
      line_count = 0
      occ_housing_difference = []
      temp_ohv = 0.00
      diff_ohv = 0.00
      value = 0
      file_data_list = []
      for row in csv_reader:
        if line_count == 0:
          print(f'Data column names are {", ".join(row)}')
          line_count += 1
        else:
          file_data_list.append(row)
          if row[1] == "Occupied Housing Units":
            value = row[2].replace(" ","")
            if temp_ohv == 0:
              temp_ohv = float(value)
            else:
              diff_ohv = temp_ohv-float(value)
              temp_ohv = float(value)
              occ_housing_difference.append(diff_ohv)
      occ_housing_difference.sort()
      csv_file.close()
      print("\nQuestion no.: 4.(a)")
      print("Consecutive year differences of Occupied houses in ascending order are : \n", occ_housing_difference)
  else:
    print("No such file is found")
  return file_data_list 

#Using function to calculate percentage of Vacant and Occupied homes and printing results in given format
def house_unit_percent(file_data):
  print("\nQuestion no.: 4.(b) after rounding up the values")
  print('Year\tVacant Housing Units\tOccupied Housing Units')
  final_dict = {}
  for i in range(len(file_data)):
    if file_data[i][1] == "Total Housing Units":
      total_houses = float(file_data[i][2].replace(" ",""))
      year = file_data[i][0]
      final_dict[year] = [total_houses]
        
    elif file_data[i][1] == "Vacant Housing Units" and file_data[i][0] == year:
      v_percent = 100*(float(file_data[i][2].replace(" ",""))/total_houses)
      final_dict[year].append(v_percent)
        
    elif file_data[i][1] == "Occupied Housing Units" and file_data[i][0] == year:
      o_percent = 100*(float(file_data[i][2].replace(" ",""))/total_houses)
      final_dict[year].append(o_percent)
    
  for key,value in final_dict.items():
    print('\n'+key+'\t'+str(value[1])+'%\t'+str(value[2])+'%')

  print("\nQuestion no.: 4.(b) after rounding up the values")
  print('Year\tVacant Housing Units\tOccupied Housing Units')
  for key,value in final_dict.items():
    print('\n'+key+'\t'+str(round(value[1]))+'%\t\t\t'+str(round(value[2]))+'%')
  return

#Using function to calculate highest Vacant and Occupied homes from all years and printing results in given format
def highest_housing_units(file_data):
  h_vacant_houses = 0
  h_occupied_houses = 0
  h_v_list = []
  h_o_list = []

  for each_row in file_data:
    if each_row[1] == "Vacant Housing Units":
      value = float(each_row[2].replace(" ",""))
      if h_vacant_houses == value:
        h_v_list.append([each_row[0],h_vacant_houses])
      elif h_vacant_houses < value:
        h_vacant_houses = value
        h_v_list.clear()
        h_v_list.append([each_row[0],h_vacant_houses])

    elif each_row[1] == "Occupied Housing Units":
      value = float(each_row[2].replace(" ",""))
      if h_occupied_houses == value:
        h_o_list.append([each_row[0],h_occupied_houses])
      elif h_occupied_houses < value:
        h_occupied_houses = value
        h_o_list.clear()
        h_o_list.append([each_row[0],h_occupied_houses])

  print("\nQuestion no.: 4.(c)")    
  print('\t\t\tYear\tValue')
  for each in h_v_list:
    print(f'Vacant Housing Units:\t{each[0]}\t{each[1]}')
  for each in h_o_list:
    print(f'Occupied Housing Units:\t{each[0]}\t{each[1]}')
  return

#Using function to calculate Harmonic Mean of Total Housing Units and printing results
def harmonic_total_houses(file_data):
  total_sum = 0 
  denom_sum = 0
  for each_row in file_data:
    if each_row[1] == "Total Housing Units":
      total_sum +=1
      denom_sum += 1/int(each_row[2].replace(" ",""))
  
  return total_sum/denom_sum

#Calling above function to open and read csv file, print results from difference and storing file content in rows  
files_data = diff_calc()

#Calling above function to percent of vacant and occupied homes and  printing results in given format
house_unit_percent(files_data)

#Calling above function to calculate highest vacant and occupied homes with year and  printing results in given format
highest_housing_units(files_data)     

#Calling above function to calculate harmonic mean of Total Housing units and printing results
print("\nQuestion no.: 4.(d)")
print("Harmonic mean of Total Housing Units: ", harmonic_total_houses(files_data))
  
```
